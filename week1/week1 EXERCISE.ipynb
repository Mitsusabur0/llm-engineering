```python
# constants

MODEL = 'gpt-5-nano'
MODEL_GEMINI = 'gemini-2.5-pro'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment

load_dotenv(override=True)
openai_api_key = os.getenv("OPENAI_API_KEY")
gemini_api_key = os.getenv("GOOGLE_API_KEY")
```

```python
system_prompt = """
You'll get the script of a YouTube video.
These videos will be about rpg character creation and builds.
Your task is to summarize the video script into a character build guide.
These guides will be for the games Baldur's Gate 3 or Pathfinder: Wrath of the Righteous.
"""
```

```python
ytt_api = YouTubeTranscriptApi()

def get_video_script(video_id):
    script = ytt_api.fetch(video_id)
    full_script = ""
    for snippet in script:
        full_script += snippet.text + " "
    return full_script
```

```python
def get_user_prompt(script):
    return f"""
        Here is the script of a YouTube video:
        {script}
        Please summarize this script into a character build guide.
    """
```

```python
script = get_video_script("FgOfzDlOQ9w")

response = openai.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": get_user_prompt(script)}
    ],
)
print(response.choices[0].message.content)
```
